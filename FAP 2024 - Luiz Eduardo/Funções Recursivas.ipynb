```python
def somatorio(n):
  if n == 0:
    return 0
  else:
    return n + somatorio(n-1)

n = int(input("Digite um número: "))
print('o somatório é', somatorio(n))
```

```python
def fatorial(n):
  if n == 1:
    return 1
  else:
    return n * fatorial(n-1)

n = int(input("Digite um número: "))
print(f'{n}! = ', fatorial(n))
```
