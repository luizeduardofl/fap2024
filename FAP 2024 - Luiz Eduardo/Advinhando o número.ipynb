## Importando a biblioteca random

```python
import random
```

## O usuário tenta advinhar o número gerado pela máquina

```python
# A quantidade de tentativas começa com 0.
tentativas = 0

# Limite superior para o intervalo.
lim_superior = 100

# Gerando um número inteiro pseudoaleatório.
x = random.randint(1, lim_superior)

# Pedindo um número do usuário.
y = int(input('Caro usuário, digite um número entre 1 e {}: '.format(lim_superior)))

# Tratamento de erro: obrigando o usuário a digitar um número válido.
while y > lim_superior or y < 1:
  y = int(input('Número inválido. Digite o número entre 1 e {}:'.format(lim_superior)))

# Enquanto o usuário não acertar o número, ele deve inseri-lo novamente.
while x != y:
  if x > y:
    y = int(input('Você errou. Tente novamente. Uma dica: O número é maior que {}: '.format(y)))
  else:
    y = int(input('Você errou. Tente novamente. Uma dica: O número é menor que {}: '.format(y)))
  tentativas += 1

# Impressão da mensagem de derrota caso o usuário exceda cinco tentativas.
  if tentativas > 5:
    print('Infelizmente, o número de tentativas acabou.')
    print('GAME OVER!!!')
    break

# Imprimindo a mensagem da possível vitória.
if tentativas > 1:
  print('Você acertou com {0} tentativas. Parabéns! O número era, de fato, {1}.'.format(tentativas, x))
else:
  print('Você acertou com apenas uma tentativa. Parabéns! De fato, o número era {0}.'.format(x))
```
