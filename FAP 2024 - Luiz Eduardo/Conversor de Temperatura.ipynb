## Definindo as funções

### Saudações + recebendo a temperatura do usuário

```python
def saudacao():
  print('Seja bem-vindo. Este conversor de temperatura converterá a entrada no sistema de temperatura que você desejar.')
  print('Existem três sistemas principais: Celsius, Fahrenheit e Kelvin (°C, °F ou K).')

def recebe_temperatura():
  temp = float(input('Digite a temperatura que deseja converter: '))
  return round(temp, 2)
```

### Convertendo a temperatura e armazenando as escalas inseridas

```python
def recebe_escalas():
  escala1 = input('Digite o sistema da temperatura que você inseriu anteriormente (°C, °F ou K): ').upper()
  while escala1 not in ['C', 'F', 'K']:
    escala1 = input(('Entrada inválida. Digite um dos três sistemas possíveis.')).upper()

  escala2 = input('Agora, digite o sistema de temperatura para o qual você deseja converter (°C, °F ou K): ').upper()
  while escala2 not in ['C', 'F', 'K']:
    escala2 = input(('Entrada inválida. Digite um dos três sistemas possíveis.')).upper()

  return escala1, escala2

def conversor_temperatura(temp, escala_entrada, escala_saida):
  if escala_entrada == 'C':
    match escala_saida:
      case 'F':
       temp_nova = round((temp * 9/5) + 32, 1)
      case 'K':
       temp_nova = round(temp + 273.15, 1)
  if escala_entrada == 'F':
    match escala_saida:
      case 'C':
       temp_nova = round((temp - 32) * 5/9, 1)
      case 'K':
       temp_nova = round((temp + 459.67) * 5/9, 1)
  if escala_entrada == 'K':
    match escala_saida:
      case 'C':
       temp_nova = round(temp - 273.15, 1)
      case 'F':
       temp_nova = round((temp * 9/5) - 459.67, 1)

  return round(temp_nova, 2)
```

## Conversor de Temperatura com três escalas (Celsius, Fahrenheit e Kelvin)

```python
saudacao()

temp_usuario = recebe_temperatura()

escala_entrada, escala_saida = recebe_escalas()

temp_nova = conversor_temperatura(temp_usuario, escala_entrada, escala_saida)
```

## Impressão na tela

```python
if escala_saida != 'K':
  print('A temperatura nova é: {} °{}.'.format(temp_nova, escala_saida))
else:
  print('A temperatura nova é: {} K.'.format(temp_nova))
```
