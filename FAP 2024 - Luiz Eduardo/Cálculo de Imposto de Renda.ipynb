## Importando as bibliotecas

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Definindo as funes


```python
def saudacao():
  print('Bem-vindo  calculadora simplificada de imposto de renda. \n')

def calcular_imposto(salario):
  if (salario <= 2259.20):
    return (salario * 0 - 0)
  if (salario > 2259.20 and salario <= 2826.65):
    return (salario * 0.075 - 169.44)
  if (salario > 2826.65 and salario <= 3751.05):
    return (salario * 0.15 - 381.44)
  if (salario > 3751.05 and salario <= 4664.68):
    return (salario * 0.225 - 662.77)
  if (salario > 4664.68):
    return (salario * 0.275 - 896)

def validacao_salario(salario):
  while (salario < 1412):
    salario = float(input('Por favor, digite um salrio vlido. O salrio mnimo em 2024  de R$ 1412.'))

def mensagem_imposto(salario, imposto):
  print("Seu salrio vale R$ {} e o imposto de renda incidente  de R$ {}.".format(salario, imposto))
  print('Portanto, seu salrio lquido vale R$ {}.'.format(salario-imposto))

def calcular_salario_liquido(salario, imposto):
  return (salario - imposto)
```

## Usando a calculadora com quatro casos hipotticos

```python
saudacao()
casos = [2300, 3000, 4100, 5200]
impostos_casos = []
for salario in casos:
  imposto = calcular_imposto(salario)
  impostos_casos.append(round(imposto, 2))

for i in range(4):
  mensagem_imposto(casos[i], impostos_casos[i])
  print('\n')
```

## Explicando graficamente a tabela

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 6))


# x representa o salrio bruto; y, o imposto de renda; z, o salrio lquido.
x = np.linspace(0, 20000, 20)
y = []
for salario in x:
  y.append(calcular_imposto(salario))
z = []
for salario in x:
  z.append(calcular_salario_liquido(salario, calcular_imposto(salario)))

# Primeiro sub-grfico
ax1.plot(x, y, color='red')
ax1.set_xlabel('Salrio bruto')
ax1.set_ylabel('Imposto de renda')
ax1.set_title('Imposto de renda em funo do salrio bruto (em reais)')
ax1.set_ylim(0, 5000)


# Segundo sub-grfico
ax2.plot(x, z, color='green')
ax2.set_xlabel('Salrio bruto')
ax2.set_ylabel('Salrio lquido')
ax2.set_title('Salrio lquido em funo do salrio bruto (em reais)')
ax2.set_ylim(0, 20000)

# Ajuste de layout
plt.tight_layout()

# Mostrar o grfico
plt.show()
```

Com o primeiro sub-grfico acima, percebemos visualmente que o imposto de renda  feito para ser menos agressivo a salrios brutos menores - principalmente aos menores que R$ 2259,20. A partir desse valor, o valor do imposto aumenta linearmente em relao ao salrio bruto.

Como consequncia da linearidade de Imposto de renda x salrio bruto, o segundo sub-grfico tambm revela um comportamento linear.
