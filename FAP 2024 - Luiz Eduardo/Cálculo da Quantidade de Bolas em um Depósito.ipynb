## Importando a biblioteca math

```python
import math
```

## Definindo funções

### Saudação + funções que determinam volumes

```python
def saudacao():
  print('Você dispõe de inúmeras bolas em sua loja e não sabe quantas cabem no seu depósito? Esta calculadora o ajudará nisso.')

def det_volume_bola(diametro):
  volume = (4/3) * math.pi * (diametro/2) ** 3
  return round(volume, 2)

def det_volume_deposito():
  print('Agora vamos pedir as dimensões do seu depósito em centímetros. Responda atentamente às perguntas.')

  comprimento = float(input('Digite o comprimento do depósito em centímetros: '))
  while comprimento == 0:
    comprimento = float(input('Por favor, digite um valor de comprimento válido: '))

  largura = float(input('Digite a largura do depósito em centímetros: '))
  while largura == 0:
    largura = float(input('Por favor, digite um valor de largura válido: '))

  altura = float(input('Digite a altura do depósito em centímetros: '))
  while altura == 0:
    altura = float(input('Por favor, digite um valor de altura válido: '))

  volume = comprimento * largura * altura
  return round(volume, 2)

def det_bolas_deposito(volume_deposito, volume_bola):
  bolas_deposito = volume_deposito / volume_bola
  return int(bolas_deposito)
```

### Bola escolhida pelo usuário

```python
def opcoes_bola():
  print('Caro usuário, escolha uma das opções disponíveis abaixo ou pressione 7 para inserir um tamanho personalizado.')
  print('1 - Bola de Basquete Adulto (24 cm)')
  print('2 - Bola de Basquete Infantil (22 cm)')
  print('3 - Bola de Futebol Oficial (22 cm)')
  print('4 - Bola de Vôlei (21 cm)')
  print('5 - Bola de Futebol de Salão (20 cm)')
  print('6 - Bola de Handball (19 cm)')
  print('7 - Tamanho personalizado')
  bola_escolhida = int(input())

  while not bola_escolhida in range(1, 8):
    bola_escolhida = int(input('Por favor, escolha um número válido: '))

  return bola_escolhida
```

##  Cálculo da Quantidade de Bolas em um Depósito

### Calculando quantas bolas cabem no depósito

```python
saudacao()
volume_deposito_usuario = det_volume_deposito()

diametro_bolas = [24, 22, 22, 21, 20, 19, None]

bola_escolhida = opcoes_bola()

diametro_bola_usuario = diametro_bolas[bola_escolhida - 1]

if bola_escolhida == 7:
  diametro_bola_usuario = float(input("Você escolheu a opção 'Tamanho personalizado'. Agora, digite o diâmetro da bola em centímetros: "))

  while diametro_bola_usuario == 0:
    diametro_bola_usuario = float(input('Por favor, digite um valor de diâmetro válido: '))

  diametro_bolas[6] = diametro_bola_usuario

volume_bola_usuario = det_volume_bola(diametro_bola_usuario)

numero_bolas_deposito = det_bolas_deposito(volume_deposito_usuario, volume_bola_usuario)
```

### Imprimindo mensagem na tela

```python
if numero_bolas_deposito < 2:
  print('O volume do seu depósito é {0} cm³. O volume da bola escolhida é {1} cm³. Portanto, cabe {2} bola no depósito.'
.format(volume_deposito_usuario, volume_bola_usuario, numero_bolas_deposito))

else:
  print('O volume do seu depósito é {0} cm³. O volume da bola escolhida é {1} cm³. Portanto, cabem {2} bolas no depósito.'
.format(volume_deposito_usuario, volume_bola_usuario, numero_bolas_deposito))
```
