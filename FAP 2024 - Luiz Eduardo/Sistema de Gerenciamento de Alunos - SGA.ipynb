## Definindo as funções

```python
def apresentacao(uso):
  if uso == 0:
    print('Seja bem-vindo, caro usuário. ')
    print('Este é o Sistema de Gerenciamento de Alunos (SGA). Aqui, diversas operações são possíveis: ')
    print('1 - Cadastrar novo aluno')
    print('2 - Pesquisar aluno')
    print('3 - Remover aluno')
    print('4 - Editar aluno')
    print('5 - Calcular média de notas de um aluno')
    print('6 - Listar todos os alunos cadastrados')
    print('7 - Apagar TODOS os alunos [RESTRITO]')
    print('8 - Sair')
    print('Vamos começar.\n')

  else:
    print('\nAgora que sua operação foi concluída, deseja fazer algo mais? ')
    print('1 - Cadastrar novo aluno')
    print('2 - Pesquisar aluno')
    print('3 - Remover aluno')
    print('4 - Editar aluno')
    print('5 - Calcular média de notas de um aluno')
    print('6 - Listar todos os alunos cadastrados')
    print('7 - Apagar TODOS os alunos [RESTRITO]')
    print('8 - Sair')


def cadastrar_aluno():
  print('Você escolheu a opção 1 - Cadastrar novo aluno')
  dic = {}

  dic['nome'] = input('Digite o nome do aluno: ').title()
  while len(dic['nome']) < 3:
    dic['nome'] = input('Nome inválido. Digite o nome do aluno novamente: ').title()

  dic['curso'] = input('Digite o curso do aluno: ').title()
  while len(dic['curso']) < 3:
    dic['curso'] = input('Curso inválido. Digite o nome do curso novamente: ').title()

  dic['matrícula'] = input('Digite a matrícula do aluno: ')
  while not len(dic['matrícula']) == 11:
    dic['matrícula'] = input('Matrícula inválida. Digite a matrícula do aluno novamente: ')
  dic['matrícula'] = int(dic['matrícula'])

  for i in range(len(alunos)):
    while alunos[i]['matrícula'] == dic['matrícula']:
      dic['matrícula'] = int(input('Esta matrícula já existe no sistema. Digite um número diferente da atual: '))

  notas_unidades = []
  for i in range(3):
      nota = float(input('Digite a nota da unidade {}: '.format(i+1)))
      while nota not in range(11):
        nota = float(input('A nota deve estar entre 0 e 10. Digite a nota da unidade {} novamente: '.format(i+1)))
      notas_unidades.append(nota)

  dic['notas'] = notas_unidades

  alunos.append(dic)
  print('Aluno cadastrado com sucesso!\n')

def pesquisar_aluno():
  print('Você escolheu a opção 2 - Pesquisar aluno\n')

  if len(alunos) == 0:
    print('Não há nenhum aluno cadastrado. Logo, essa operação é inválida.')
    return

  print('Você pode pesquisar pelo aluno usando nome, curso ou matrícula.\n')

  resposta = input('Pesquisar por: ')
  try:
    resposta = int(resposta)
    encontrado = False
    for i in range(len(alunos)):
      if alunos[i]['matrícula'] == resposta:
        print('Aluno encontrado! {}'.format(alunos[i]))
        encontrado = True
        break

    if not encontrado:
      print('Nenhum aluno encontrado com a informação dada.')

  except ValueError as ve:
    resposta = resposta.title()

    encontrado = False
    for i in range(len(alunos)):
      if alunos[i]['nome'] == resposta or alunos[i]['curso'] == resposta:
        print('Aluno encontrado! {}'.format(alunos[i]))
        encontrado = True
        break

    if not encontrado:
      print('Nenhum aluno encontrado.')

def remover_aluno():
  print('Você escolheu a opção 3 - Remover aluno\n')

  if len(alunos) == 0:
    print('Não há nenhum aluno cadastrado. Logo, essa operação é inválida.')
    return

  print('Só é possível remover um aluno usando sua matrícula, pois ela é única para cada um.\n')

  resposta = int(input('Insira a matrícula: '))
  encontrado = False
  for i in range(len(alunos)):
    if alunos[i]['matrícula'] == resposta:
      print('O seguinte aluno foi encontrado: {}'.format(alunos[i]))
      encontrado = True

      comando = input('Você tem certeza desta operação? [S/N] ')
      while comando not in ['S', 'N']:
        comando = input('Digite um comando válido [S/N]: ')

      if comando == 'S':
        del alunos[i]
        print('Aluno removido com sucesso!\n')
      else:
        print('Operação cancelada.\n')

      break

  if not encontrado:
      print('Nenhum aluno encontrado com essa matrícula.')
      return

def editar_aluno():
  print('Você escolheu a opção 4 - Editar aluno\n')

  if len(alunos) == 0:
    print('Não há nenhum aluno cadastrado. Logo, essa operação é inválida.')
    return

  print('Só é possível editar um aluno usando sua matrícula, pois ela é única para cada um.\n')

  resposta = int(input('Insira a matrícula: '))
  encontrado = False
  for i in range(len(alunos)):
    if alunos[i]['matrícula'] == resposta:
      print('O seguinte aluno foi encontrado: {}'.format(alunos[i]))
      encontrado = True

      comando = input('Que atributo de {} você gostaria editar: nome, curso, matrícula ou notas?'.format(alunos[i]['nome']))
      comando = comando.lower()

      while comando not in ['nome', 'curso', 'matrícula', 'notas']:
        comando = input('Digite um comando válido [nome/curso/matrícula/notas]: ')

      if comando == 'nome':
        alunos[i]['nome'] = input('Digite o novo nome: ').title()
        while len(alunos[i]['nome']) < 3:
          alunos[i]['nome'] = input('Só serão aceitos nomes com pelo menos três caracteres. Digite um novo nome: ').title()
      print('Nome editado com sucesso!\n')

      if comando == 'curso':
        alunos[i]['curso'] = input('Digite o novo curso: ').title()
        while len(alunos[i]['curso']) < 3:
          alunos[i]['curso'] = input('Só serão aceitos cursos com pelo menos três caracteres. Digite um novo curso: ').title()
      print('Curso editado com sucesso!\n')

      if comando == 'matrícula':
        alunos[i]['matrícula'] = int(input('Digite a nova matrícula: '))
        while not len(alunos[i]['matrícula']) == 11:
          alunos[i]['matrícula'] = int(input('A matrícula deve conter 11 dígitos. Digite a nova matrícula: '))
        print('Matrícula editada com sucesso!\n')

      if comando == 'notas':
        alunos[i]['notas'] = []
        for j in range(3):
          nota = float(input('Digite a nova nota da unidade {}: '.format(j+1)))
          while nota not in range(11):
            nota = float(input('A nota deve estar entre 0 e 10. Digite a nota da unidade {} novamente: '.format(j+1)))
          alunos[i]['notas'].append(nota)
        print('Notas editadas com sucesso!\n')

      break

  if not encontrado:
      print('Nenhum aluno encontrado com essa matrícula.')
      return

def media_aluno():
  print('Você escolheu a opção 5 - Calcular média de notas de um aluno\n')

  if len(alunos) == 0:
    print('Não há nenhum aluno cadastrado. Logo, essa operação é inválida.')
    return

  print('Só é possível calcular a média de notas de um aluno usando sua matrícula, pois ela é única para cada um.\n')

  resposta = int(input('Insira a matrícula: '))
  encontrado = False
  for i in range(len(alunos)):
    if alunos[i]['matrícula'] == resposta:
      print('O seguinte aluno foi encontrado: {}'.format(alunos[i]))
      encontrado = True
      break

  if not encontrado:
      print('Nenhum aluno encontrado com essa matrícula.')
      return

  notas = alunos[i]['notas']
  media = round(statistics.mean(notas), 2)

  print('A média de notas de {0} é {1}'.format(alunos[i]['nome'], media))



def listar_alunos():
  print('Você escolheu a opção 6 - Listar todos os alunos cadastrados\n')

  if len(alunos) > 0:
    for i in range(len(alunos)):
      print('Aluno número {0}: {1}'.format(i+1, alunos[i]))
      print('______________________________________________\n')
  else:
      print('Não há alunos cadastrados. Logo, essa operação é inválida.')

def apagar_dados():
  print('Você escolheu a opção 7 - Apagar TODOS os alunos [RESTRITO]')
  print('Essa operação é irreversível. Use-a com cautela.\n')

  senha = '/suporte'
  tentativas = 3

  senha_usuario = input('Digite a senha para autorizar a operação: ')
  while senha_usuario != senha:
    tentativas -= 1
    senha = input('Tente novamente para autorizar a operação: ')
    if tentativas == 0:
        print('Seu número de tentativas acabou. Tente novamente mais tarde.')
        break

  global alunos
  alunos = []
  print('Todos os alunos foram apagados com sucesso.\n')
```

## Programa propriamente dito

```python
alunos = []
uso = 0

while True:
  apresentacao(uso)

  opcao = int(input('Escolha uma opção: '))
  while opcao not in range(1, 9):
    opcao = int(input('Sua opção não é permitida. Escolha um número válido, por favor: '))

  if opcao == 1:
    cadastrar_aluno()

  if opcao == 2:
    pesquisar_aluno()

  if opcao == 3:
    remover_aluno()

  if opcao == 4:
    editar_aluno()

  if opcao == 5:
    media_aluno()

  if opcao == 6:
    listar_alunos()

  if opcao == 7:
    apagar_dados()

  if opcao == 8:
    print('Obrigado por usar o SGA. Até a próxima!')
    break

  if uso > 4:
    print('\nCaro usuário, seu limite diário foi atingido (5 usos).')
    print('Para ter acesso ilimitado, assine o SGA Plus por apenas R$ 9,90 por mês!')
    break

  uso += 1

print('\nO SGA foi desenvolvido por Luiz Lobato em 2024. Para saber mais, entre em contato pelo e-mail: ')
print('luizeduardofl11@gmail.com')
```
