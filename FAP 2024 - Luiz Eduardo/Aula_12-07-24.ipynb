```python
class Pessoa:
  def __init__(self, nome, idade, emprego):
    self.nome = nome
    self.idade = idade
    self.emprego = emprego

  def __add__(self, other):
    print('{0} e {1} são bons demais. Parabéns, galera! Valeu.'.format(self.nome, other.nome))

  def imprime_idade(self):
    print('{0} tem {1} anos de idade.'.format(self.nome, self.idade))

  def imprime_emprego(self):
    print('{0} trabalha como {1}.'.format(self.nome, self.emprego))

  def mensagem_motivacional(self):
    if self.nome == 'Silvio Santos':
      print('Olhe o aviãozinho! Mah oi!')
    elif self.nome == 'Ricardo':
      print('Pô, rapaz, Linux é de lascar. Vá pro Windows, homem!'.format(self.nome))
    else:
      print('{} é o cara! Show de bola!'.format(self.nome))
```

```python
pessoa1 = Pessoa('Luiz Eduardo', 20, 'Desenvolvedor Web')

pessoa2 = Pessoa('Ivan Varella', 41, 'Programador')

pessoa3 = Pessoa('Ricardo', 45, 'operador de Linux')

pessoa4 = Pessoa('José Alfredo', 54, 'professor de Engenharia Elétrica')

pessoa5 = Pessoa('Silvio Santos', 93, 'apresentador de televisão')
```

```python
lista_pessoas = [pessoa1, pessoa2, pessoa3, pessoa4, pessoa5]

for i in lista_pessoas:
  i.imprime_idade()
  i.imprime_emprego()
  i.mensagem_motivacional()
  print()
```
