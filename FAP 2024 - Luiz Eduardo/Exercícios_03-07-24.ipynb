# Questão 6 - classificador de turma por idade

```python
import statistics as st

print('Bem-vindo ao classificador de turma por idade.')
print('Vamos coletar a idade do número de pessoas que você desejar e vamos classificar a turma pela média de idade.')
print('Para finalizar a coleta, basta digitar zero (0). \n')

idade_alunos = []
idade = int(input('Para começar, digite a idade do 1º aluno: '))
quantidade_alunos = 0
while idade != 0:
    idade_alunos.append(idade)
    idade = int(input('Digite a idade do próximo aluno: '))
    quantidade_alunos += 1

if quantidade_alunos != 0:
  media_idade = st.mean(idade_alunos)

else:
  print('A turma não tem alunos.')

if media_idade <= 25:
  categoria = 'Jovem'

if media_idade > 25 and media_idade <= 60:
  categoria = 'Adulta'

if media_idade > 60:
  categoria = 'Idosa'

if quantidade_alunos > 1:
  print('A turma tem {0} alunos e a média de idade é {1}. Logo, ela é considerada {2}.'.format(quantidade_alunos, media_idade, categoria))

if quantidade_alunos == 1:
  print('A turma tem {0} aluno e a média de idade é {1}. Logo, ela é considerada {2}.'.format(quantidade_alunos, media_idade, categoria))
```

# Questão 8 - sistema de votação

```python
import matplotlib.pyplot as plt

votos_candidato_a = 0
votos_candidato_b = 0
votos_candidato_c = 0
votos_candidato_d = 0
votos_nulos = 0
votos_brancos = 0
votos_totais = 0

print('Bem-vindo sistema de votação! Neste exemplo, há quatro candidatos:')
print('Candidato A - digite A ou a para votar.')
print('Candidato B - digite B ou b para votar.')
print('Candidato C - digite C ou c para votar.')
print('Candidato D - digite D ou d para votar.')

print('Ainda há outras duas opções de voto: ')
print('Voto nulo - digite N ou n')
print("Voto em branco - digite 'BRANCO', 'Branco' ou 'branco' ")

print('Para finalizar a votação, basta digitar zero (0).')

while True:
  voto = input('Digite o seu voto: ')

  if voto == 'A' or voto == 'a':
    votos_candidato_a += 1
    votos_totais += 1

  if voto == 'B' or voto == 'b':
    votos_candidato_b += 1
    votos_totais += 1

  if voto == 'C' or voto == 'c':
    votos_candidato_c += 1
    votos_totais += 1

  if voto == 'D' or voto == 'd':
    votos_candidato_d += 1
    votos_totais += 1

  if voto == 'N' or voto == 'n':
    votos_nulos += 1
    votos_totais += 1

  if voto in ['BRANCO', 'Branco', 'branco']:
    votos_brancos += 1
    votos_totais += 1

  if voto == '0':
    break

porcentagem_candidato_a = round((votos_candidato_a / votos_totais) * 100, 2)
porcentagem_candidato_b = round((votos_candidato_b / votos_totais) * 100, 2)
porcentagem_candidato_c = round((votos_candidato_c / votos_totais) * 100, 2)
porcentagem_candidato_d = round((votos_candidato_d / votos_totais) * 100, 2)
porcentagem_votos_brancos = round((votos_brancos / votos_totais) * 100, 2)
porcentagem_votos_nulos = round((votos_nulos / votos_totais) * 100, 2)

if votos_candidato_a > 1:
  print('Cadidato A - {0} votos ({1}%)'.format(votos_candidato_a, porcentagem_candidato_a))
else:
  print('Cadidato A - {0} voto ({1}%)'.format(votos_candidato_a, porcentagem_candidato_a))

if votos_candidato_b > 1:
  print('Cadidato B - {0} votos ({1}%)'.format(votos_candidato_b, porcentagem_candidato_b))
else:
  print('Cadidato B - {0} voto ({1}%)'.format(votos_candidato_b, porcentagem_candidato_b))

if votos_candidato_c > 1:
  print('Cadidato C - {0} votos ({1}%)'.format(votos_candidato_c, porcentagem_candidato_c))
else:
  print('Cadidato C - {0} voto ({1}%)'.format(votos_candidato_c, porcentagem_candidato_c))

if votos_candidato_d > 1:
  print('Cadidato D - {0} votos ({1}%)'.format(votos_candidato_d, porcentagem_candidato_d))
else:
  print('Cadidato D - {0} voto ({1}%)'.format(votos_candidato_d, porcentagem_candidato_d))

print('Votos em branco - {0} ({1}%)'.format(votos_brancos, porcentagem_votos_brancos))
print('Votos nulos - {0} ({1}%)'.format(votos_nulos, porcentagem_votos_nulos))

opcoes_voto = ['Candidato A', 'Candidato B', 'Candidato C', 'Candidato D', 'Votos em branco', 'Votos nulos']
porcentagens = [porcentagem_candidato_a, porcentagem_candidato_b, porcentagem_candidato_c, porcentagem_candidato_d, porcentagem_votos_brancos, porcentagem_votos_nulos]

opcoes_voto_filtradas = []
porcentagens_filtradas = []
for i in range(len(porcentagens)):
  if porcentagens[i] != 0:
    opcoes_voto_filtradas.append(opcoes_voto[i])
    porcentagens_filtradas.append(porcentagens[i])

fig, ax = plt.subplots()
ax.pie(porcentagens_filtradas, labels=opcoes_voto_filtradas, autopct='%1.2f%%')
ax.axis('equal')

if votos_totais > 1:
  ax.set_title("Resultados da votação ({} votos no total)".format(votos_totais))
else:
  ax.set_title("Resultados da votação ({} voto no total)".format(votos_totais))

plt.show()
print('Obrigado pela participação!')
```

# Questão 12 - encriptador de três letras (Cifra de César com troca de três posições)

```python
print('Este é o encriptador baseado na Cifra de César.')
print('Digite um texto e ele será criptografado. \n')
print("Atenção: não é permitido inserir caracteres com acento ('ã', 'ê', 'í' e assim por diante.) \n")

texto = input('Digite o texto: ')

texto_criptografado = ''

for letra in texto:
  if letra == 'x':
    texto_criptografado += 'a'
  elif letra == 'y':
    texto_criptografado += 'b'
  elif letra == 'z':
    texto_criptografado += 'c'
  else:
    if letra == ' ':
      texto_criptografado += ' '
    else:
      texto_criptografado += chr(ord(letra) + 3)

print('\nTexto criptografado: {}'.format(texto_criptografado))

```

# Questão 14 - compressor de strings

```python
print('Este é o compressor de strings.')
print('Digite um texto e ele será comprimido. \n')
print('Veja o exemplo abaixo: ')
print('Antes: aabbbccccc')
print('Depois: 2a3b5c \n')

texto = input('Digite o texto: ')
texto_comprimido = ''

for letra in texto:
  if letra in texto_comprimido:
    continue

  i = 1
  while letra * i in texto:
    i += 1

  texto_comprimido += str(i-1) + letra


print('\nTexto comprimido: {}'.format(texto_comprimido))
```

# Questão 16 - contador de linhas, palavras e caracteres de um arquivo de texto

```python
print('Este é o contador de linhas, palavras e caracteres de um arquivo de texto.')
print('Usarei como exemplo um arquivo .txt da pasta que contém este notebook.\n')

def conta_linhas():
  numero_linhas = 0
  with open('/content/Texto_teste.txt', 'r') as arquivo:
    for linha in arquivo:
      numero_linhas += 1
      return numero_linhas


def conta_palavras():
  numero_palavras = 0
  with open('/content/Texto_teste.txt', 'r') as arquivo:
    for linha in arquivo:
      palavras = linha.split()
      numero_palavras += len(palavras)
      return numero_palavras

def conta_caracteres():
  numero_caracteres = 0
  with open('/content/Texto_teste.txt', 'r') as arquivo:
    for linha in arquivo:
      numero_caracteres += len(linha)
    return numero_caracteres

print('Número de linhas: {}'.format(conta_linhas()))
print('Número de palavras: {}'.format(conta_palavras()))
print('Número de caracteres: {}'.format(conta_caracteres()))
```
