```python
add = lambda x, y: x + y

result = add(3, 4)

print(result)

# Output: 7
```

```python
def add (x, y):
    return x + y

result = add(3, 4)

print(result)

# Output: 7
```

```python
def comprimento(n):
  return len(n)


frutas = list(map(comprimento, ('apple', 'banana', 'cherry')))

for i in frutas:
  print(i)
```

```python
def soma(a,b):
  return a + b

x = list(map(soma, ('apple', 'banana', 'cherry'), ('orange', 'lemon', 'pineapple')))

print(x)
```

```python
from functools import reduce

vendas = [30, 70, 100, 80, 120]
soma = reduce(lambda x, y: x + y, vendas)

media = soma / len(vendas)

print(media)
```
